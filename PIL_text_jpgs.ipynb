```python
from PIL import Image
from PIL import ImageFont, ImageDraw
import numpy as np
import random
import os
import re
import glob
```

```python
# texts = ["Hello world"]

def create_text_img(imsize: tuple, texts: list):
  # image opening
  w, h = imsize
  font_size = w//8 if h > w else h//8 
  font_size = random.randint(font_size*0.5, font_size*1.5)
  
  fnt = ImageFont.truetype("/Library/Fonts/Arial Unicode.ttf", int(font_size/2))

  txt_color = (random.randint(0,255), random.randint(0,255), random.randint(0,255))

  txt = Image.new("RGB", imsize, (255, 255, 255))
  d = ImageDraw.Draw(txt)
  text_w, text_h = random.randint(w*.25, w*.75), random.randint(h*.25, h*.75)
  for text in texts:
    d.text((text_w, text_h) , text, font=fnt, fill=txt_color)
    text_h += font_size//2
    text_w += random.randint(-font_size//4, font_size//4)
  # print(full_path)
  # out.save(output_image_path)
  return txt
```

```python
def add_text(im, texts, angle = 0):
  # image opening
  w, h = im.size
  font_size = w//8 if h > w else h//8 
  font_size = random.randint(font_size*0.5, font_size*2.5)
  wim = im.copy().rotate(-angle, expand=True).convert("RGBA") #watermark image
  fnt = ImageFont.truetype("/Library/Fonts/Arial Unicode.ttf", int(font_size/2))

  txt_color = (random.randint(0,255), random.randint(0,255), random.randint(0,255), 255)

  txt = Image.new("RGBA", wim.size, (255, 255, 255, 0))
  d = ImageDraw.Draw(txt)
  text_w, text_h = random.randint(w*.25, w*.75), random.randint(h*.25, h*.75)
  for text in texts:
    d.text((text_w, text_h) , text, font=fnt, fill=txt_color)
    text_h += font_size//2
    text_w += random.randint(-font_size//4, font_size//4)
  out = Image.alpha_composite(wim, txt)
  out = out.rotate(angle, expand = True)
  wp, hp = out.size
  sw, sh = (wp-w)//2, (hp-h)//2
  out = out.crop(box = (sw, sh, w+sw, h+sh)).convert("RGB")
  # print(full_path)
  # out.save(output_image_path)
  return out
```

```python

# fname = "/Users/narju/Documents/code/educative/retouch/add_watermark_PIL/obj2__0.png"
sourcedir = "/Users/narju/Documents/code/data/coil-100/" 
targetdir = r"/Users/narju/Documents/code/educative/retouch/originals w text"
# fname = "obj2__0.png"

fnames = glob.glob(sourcedir + "*.png")
for fname in fnames:
  # print(fname)
  start = random.randint(0,len(texts)-4)
  texts_for_this_image = texts[start: start+random.randint(2,4)]
  input_image_path = os.path.join(sourcedir, os.path.basename(fname))
  im = Image.open(input_image_path)
  out = add_text(im, texts=texts_for_this_image)
  output_image_path = os.path.join(targetdir, "texted_" +os.path.basename(fname))
  out.save(output_image_path)

print("done! ")
```
